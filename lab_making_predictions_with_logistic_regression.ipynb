**5. Create a logistic regression model to predict this variable from the cleaned data.**

```python
# We are going to convert the 'film_id', 'actor_id' and 'category_id' columns in strings as they 
# are categorical varibles and should not be treated as integers.
int_to_cat=['film_id', 'actor_id','category_id']
data[int_to_cat]=data[int_to_cat].astype(str)
data.info() # to check that the previous columns were converted to string.
```

```python
# data=data.drop(columns='rental_date',axis=1) # Should i drop "rental_date" as it defines the target variable?
```

```python
import seaborn as sb

numerical_col=['rental_rate','length']
correlation_num=data[numerical_col].corr()
sb.heatmap(correlation_num, annot=True)
plt.show()
```

```python
y=data['target'] # imbalanced data
y.value_counts() 
```

```python
y=data['target']
X=data.drop(columns='target')
#X_num=data[numerical_col]
#X_cat=data.drop(columns=numerical_col,axis=1)
#X_num.info()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
X_train_num=X_train[numerical_col]
X_train_num.head()


X_test_num=X_test[numerical_col]
```

```python
X_train_cat=X_train.drop(columns=numerical_col)
X_train_cat.head()

X_test_cat=X_test.drop(columns=numerical_col)
```

```python
StandardScaler = StandardScaler().fit(X_train_num)
x_train_norm = StandardScaler.transform(X_train_num)
print(x_train_norm.shape)
x_train_norm = pd.DataFrame(x_train_norm,columns=X_train_num.columns)
x_train_norm.head(2)
```

```python
encoder = OneHotEncoder(drop='first').fit(X_train_cat)
encoded_train = encoder.transform(X_train_cat).toarray()
cols = encoder.get_feature_names(input_features=X_train_cat.columns)
onehot_encoded_train = pd.DataFrame(encoded_train, columns=cols)
onehot_encoded_train.head(2)
```

```python
# concatenating DataFrames
X_train_concat = pd.concat([x_train_norm, onehot_encoded_train], axis=1)
X_train_concat.head()
```

```python
model = LogisticRegression() 
model.fit(X_train_concat,y_train)
```

```python
encoded_test = encoder.transform(X_test_cat).toarray()
cols = encoder.get_feature_names(input_features=X_test_cat.columns)
onehot_encoded_test = pd.DataFrame(encoded_test, columns=cols)
onehot_encoded_test.head(2)
```

```python
x_test_norm = StandardScaler.transform(X_test_num) # check "transform"
print(x_test_norm.shape)
x_test_norm = pd.DataFrame(x_test_norm,columns=X_test_num.columns)
x_test_norm.head()
```

```python
X_test_concat = pd.concat([x_test_norm, onehot_encoded_test], axis=1)
X_test_concat.head()
```

```python
y_pred = model.predict(X_test_concat)
print('Accuracy of the test set: {:.2f}'.format(model.score(X_test_concat, y_pred)))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(y_test, y_pred)
print(confusion_matrix)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```
